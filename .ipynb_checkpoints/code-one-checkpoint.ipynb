```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import PIL
import tensorflow as tf
from tensorflow import keras
from keras import layers, Sequential
import pickle
```

```python
import pathlib
data_dir = pathlib.Path("Dataset Draft VKR01")
```

```python
len(list(data_dir.glob('*/*.jpg')))
```

```python
leaf_images_dict = {
    'almond': list(data_dir.glob('almond/*')),
    'hibiscus': list(data_dir.glob('hibiscus/*')),
    'java plum': list(data_dir.glob('java plum/*')),
    'money plant': list(data_dir.glob('money plant/*')),
    'rose': list(data_dir.glob('rose/*'))
}
leaf_labels_dict = {
    'almond': 0,
    'hibiscus': 1,
    'java plum': 2,
    'money plant': 3,
    'rose': 4
}
```

```python
X, y = [], []

for leaf_name, images  in leaf_images_dict.items():
    for image in images:
        img = cv2.imread(str(image))
        resized_img = cv2.resize(img,(100,100))
        X.append(resized_img)
        y.append(leaf_labels_dict[leaf_name])
```

```python
X = np.array(X)
y = np.array(y)
```

```python
X.shape
```

```python
dataset_vkr01 = {
    "X":X,
    "y":y
}
```

```python
import pickle
```

```python
with open('datasets_pickle/dataset_vkr01.pickle','wb') as f:
    pickle.dump(dataset_vkr01,f)
```

```python
def get_aug1():
    model = Sequential([
        keras.layers.RandomZoom(0.1, input_shape=(100,100,3)),
        keras.layers.RandomRotation(0.4)
    ])
    return model

aug1 = get_aug1()
```

```python
X1 = aug1(X).numpy().astype("uint8")
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.imshow(X[0])
```

```python
plt.imshow(X1[0])
```

```python
X1[0]
```

```python
up_X = np.concatenate((X,X1))
up_y = np.concatenate((y,y))
```

```python
up_X.shape
```

```python
dataaug2 = Sequential([
    keras.layers.RandomZoom(0.3, input_shape=(100,100,3)),
    keras.layers.RandomRotation(0.1),
    keras.layers.RandomFlip("horizontal")
])
X1 = dataaug2(up_X).numpy().astype("uint8")
```

```python
up_X = np.concatenate((up_X,X1))
up_y = np.concatenate((up_y,up_y))
```

```python
up_X.shape,up_y.shape
```

```python
dataset_vkr01 = {
    "X":up_X,
    "y":up_y
}
```

```python
with open('datasets_pickle/after-data-aug.pickle','wb') as f:
    pickle.dump(dataset_vkr01,f)
```
